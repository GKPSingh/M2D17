```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
## Functional Method
plt.plot(x,y);
```

```python
plt.plot(x,y,'k');
```

```python
plt.plot(x,y, 'r*')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph Y vs X')
```

```python
plt.subplot(1,2, 1 )
plt.plot(x,y, 'r*');
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y vs X')

plt.subplot(1,2,2)
plt.plot(y,x,'k--');
plt.xlabel('Y')
plt.ylabel('X')
plt.title('X vs Y')
```

```python
## Object Oriented method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]);   # left, bottom, width, height

axes.plot(x,y);
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1]);   # left, bottom, width, height
axes.plot(x,y);
```

```python
fig = plt.figure()

axes = fig.add_axes([1,1,1, 1])
axes.plot(x,y);
```

```python
fig = plt.figure()

axes = fig.add_axes([1,1,1,1]);
axes.plot(x,y);
axes.set_xlabel('X Label');
axes.set_ylabel('Y Label');
axes.set_title('Title');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,1,1]);
axes2 = fig.add_axes([0.2,0.65,0.4,0.4]);

axes1.plot(x,y);
axes2.plot(x,y);
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,1,1]);
axes2  = fig.add_axes([0.79,0.25,0.3,0.3]);
axes1.plot(x,y, 'g--');
axes2.plot(y,x, 'r*');
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Y vs X')

axes2.set_xlabel('Y Label')
axes2.set_ylabel('X Label')
axes2.set_title('X vs Y')
```

```python
fig ,axes = plt.subplots()
axes.plot(x,y,'g*')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y,'r-')
axes[0].set_title('Y vs X')
axes[1].plot(y,x,'g*')
axes[1].set_title('X vs Y')
```

```python
fig = plt.figure(figsize= (3,2))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y);
```

```python
fig = plt.figure(figsize = (7,2))
axes = fig.add_axes([0,0,1,1]);

axes.plot(x,y, 'r*');
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (7,5))
axes[0].plot(x,y, 'r-');
axes[1].plot(y,x, 'g*');
plt.tight_layout()
```

```python
fig.savefig('my_picture.jpg', dpi =200)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,'r*', label = 'Y vs X');
axes.plot(y,x,'k--', label = 'X vs Y');
axes.set_title('Title');
axes.set_xlabel('X-Label');
axes.set_ylabel('Y-Label');
axes.legend(loc = 0);
```
